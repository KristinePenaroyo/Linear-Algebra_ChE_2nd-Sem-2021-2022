##Transposition
One of the fundamental operations in matrix algebra is Transposition. The transpose of a matrix is the interchanging the values of its elements over its diagonals. With this, changing the rows and columns from the original matrix will be switched. So for a matrix $K$ its transpose is denoted as $K^T$. So for example:

$$K = \begin{bmatrix} -9 & 6 & 2\\1 & 5 &8 \\ -4 & 3 & 8\end{bmatrix} $$


$$K^T = \begin{bmatrix} -9 & 1 & -4\\6 & 5 &3 \\ 2 & 8 & 8\end{bmatrix} $$

This can now be achieved programmatically by using `np.transpose()` or using the `T` method.

```python
K = np.array([
    [-9,6,2],
    [1,5, 8],
    [-4,3,8],
])
K
```

```python
KT1 = np.transpose(K)
KT1
```

```python
KT2 = K.T
KT2
```

```python
np.array_equiv(KT1, KT2)
```

##Dot Product / Inner Product
During the laboratory activity before, we try implement the same operation with matrices. In matrix dot product we are going to get the sum of products of the vectors by row-column pairs. So if we have two matrices $L$ and $M$:



$$
L = \begin{bmatrix}l_{(0,0)}&l_{(0,1)}\\ l_{(1,0)}&l_{(1,1)}\end{bmatrix}, M = \begin{bmatrix}m_{(0,0)}&m_{(0,1)}\\ m_{(1,0)}&m_{(1,1)}\end{bmatrix}
$$

"$$L\cdot M= \begin{bmatrix} l_{(0,0)}*m_{(0,0)} + l_{(0,1)}*m_{(1,0)} & l_{(0,0)}*m_{(0,1)} + l_{(0,1)}*m_{(1,1)} \\  l_{(1,0)}*m_{(0,0)} + l_{(1,1)}*m_{(1,0)} & l_{(1,0)}*m_{(0,1)} + l_{(1,1)}*m_{(1,1)}\end{bmatrix}
$$


So if we assign values to $L$ and $M$

$$
L = \begin{bmatrix}1&2\\ 0&1\end{bmatrix}, 
M = \begin{bmatrix}-1&0\\ 2&2\end{bmatrix}
$$

$$
L \cdot M= \begin{bmatrix} 1*-1 + 2*2 & 1*0 + 2*2 \\  0*-1 + 1*2 & 0*0 + 1*2 \end{bmatrix}
= \begin{bmatrix} 3 & 4 \\2 & 2 \end{bmatrix}
$$

This could be achieved programmatically using `np.dot()`, `np.matmul()` or the `@` operator.

In matrix dot products there are additional rules compared with vector dot products. Since vector dot products were just in one dimension there are less restrictions. Since now we are dealing with Rank 2 vectors we need to consider some rules:

###Rule 1: The inner dimensions of the two matrices in question must be the same.

So given a matrix $A$ with a shape of $(a,r)$ where $a$ and $r$ are any integers. If we want to do a dot product between $A$ and another matrix $R$, then matrix $R$ should have a shape of $(r,c)$ where $r$ and $c$ are any integers. So for given the following matrices:

$$
A = \begin{bmatrix}3&-7\\2&9\\-8&4\end{bmatrix},  
R = \begin{bmatrix}5&5\\8&8\\-10&-6\end{bmatrix}, 
C = \begin{bmatrix}9&9&2\\3&9&9\end{bmatrix}
$$

So in this case $A$ has a shape of $(3,2)$, $R$ has a shape of $(3,2)$ and $C$ has a shape of $(2,3)$. So the only matrix pairs that is eligible to perform dot product is matrices $A\cdot C$, or $R\cdot C$.  

```python
P = np.array([
    [-7,9,-2,0],
    [8,5,-2,8],
    [3,-1,-4,2],
])
E = np.array([
    [9,-9,1,-5],
    [-7,8,10,0],
    [8,7,-1,-4],
])

T = np.array([
    [2,0,9,-4],
    [-2,-3,6,1],
    [7,7,-3,19],
    [6,9,0,-8],
])
print(P.shape)
print(E.shape)
print(T.shape)
```

```python
P @ T
```

```python
E @ T
```

```python
T @ T
```

```python
P @ E.T
```

If you would notice the shape of the dot product changed and its shape is not the same as any of the matrices we used. The shape of a dot product is actually derived from the shapes of the matrices used. So recall matrix $A$ with a shape of $(a,b)$ and matrix $R$ with a shape of $(r,c)$, $A\cdot R$ should have a shape $(a,c)$.

3. $A\cdot(B+C) = A\cdot B + A\cdot C$
4. $(B+C)\cdot A = B\cdot A + C\cdot A$
5. $A\cdot I = A$

##Determinant
A determinant is a scalar value derived from a square matrix. The determinant is a fundamental and important value used in matrix algebra. Although it will not be evident in this laboratory on how it can be used practically, but it will be realty used in future lessons.

The determinant of some matrix $W$ is denoted as $det(W)$ or $|W|$. So let's say $W$ is represented as:

$$
W = \begin{bmatrix}w_{(0,0)}&w_{(0,1)}\\w_{(1,0)}&w_{(1,1)}\end{bmatrix}
$$

We can compute for the determinant as:

$$|W| = w_{(0,0)}*w_{(1,1)} - w_{(1,0)}*w_{(0,1)}$$

So if we have $W$ as:

$$
W = \begin{bmatrix}1&4\\0&3\end{bmatrix}, |W| = 3
$$

But you might wonder how about square matrices beyond the shape $(2,2)$? We can approach this problem by using several methods such as co-factor expansion and the minors method. This can be taught in the lecture of the laboratory but we can achieve the strenuous computation of high-dimensional matrices programmatically using Python. We can achieve this by using `np.linalg.det()`.

```python
I = np.array([
    [-19,-9],
    [7,-20],
])
np.linalg.det(I)
```

```python
Q = np.array([
    [40,-14,-9],
    [5,-3,-21],
    [-56,-3,-14],
])
np.linalg.det(Q)
```

##Inverse
The inverse of a matrix is another fundamental operation in matrix algebra. Determining the inverse of a matrix let us determine if its solvability and its characteristic as a system of linear equation â€” we'll expand on this in the nect module. Another use of the inverse matrix is solving the problem of divisibility between matrices. Although element-wise division exists but dividing the entire concept of matrices does not exists. Inverse matrices provides a related operation that could have the same concept of \"dividing\" matrices.

Now to determine the inverse of a matrix we need to perform several steps. So let's say we have a matrix $Z$:

$$Z = \begin{bmatrix}1&7\\-3&5\end{bmatrix}$$

First, we need to get the determinant of $M$.

$$|Z| = (1)(5)-(-3)(7) = 26$$

Next, we need to reform the matrix into the inverse form:

$$Z^{-1} = \frac{1}{|Z|} \begin{bmatrix} z_{(1,1)} & -z_{(0,1)} \\ -z_{(1,0)} & z_{(0,0)}\end{bmatrix}$$

So that will be:

$$Z^{-1} = \frac{1}{26} \begin{bmatrix} 5 & -7 \\ 3 & 1\end{bmatrix} = \begin{bmatrix} \frac{5}{26} & \frac{-7}{26} \\ \frac{3}{26} & \frac{1}{26}\end{bmatrix}$$

For higher-dimension matrices you might need to use co-factors, minors, adjugates, and other reduction techinques. To solve this programmatially we can use `np.linalg.inv()`.

```python
M = np.array([
    [-19,-5],
    [9,-13],
])
np.array(M @ np.linalg.inv(M), dtype=int)
```

```python
U = np.array([
    [-18,7,-25],
    [4,-196,-56],
    [-23,13,58]
])
S = np.linalg.inv(U)
S
```
